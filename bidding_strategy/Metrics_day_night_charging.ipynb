```python
import pandas as pd
from pathlib import Path

# ======================================================
# 1. Base paths
# ======================================================
BASE = Path.cwd()

outputs_minn_miso_6h = BASE / "outputs_minn_miso_6h"

input_file = outputs_minn_miso_6h / "minn_miso_isoavg_6h_price_cov_avgdiff_spread.csv"

# ======================================================
# 2. Read the file
# ======================================================
df = pd.read_csv(input_file)

# ======================================================
# 3. Convert interval_start_utc to datetime
# ======================================================
df["interval_start_utc"] = pd.to_datetime(df["interval_start_utc"])

# Extract hour
df["hour"] = df["interval_start_utc"].dt.hour

# ======================================================
# 4. Filter intervals
# ======================================================

# File 1 → only 00:00 and 18:00
df_00_18 = df[df["hour"].isin([0, 18])].copy()

# File 2 → only 06:00 and 12:00
df_06_12 = df[df["hour"].isin([6, 12])].copy()

# Drop helper column
df_00_18.drop(columns="hour", inplace=True)
df_06_12.drop(columns="hour", inplace=True)

# ======================================================
# 5. Save CSV files
# ======================================================
df_00_18.to_csv(outputs_minn_miso_6h / "miso_intervals_00_18.csv", index=False)
df_06_12.to_csv(outputs_minn_miso_6h / "miso_intervals_06_12.csv", index=False)

print("CSV files created successfully.")
```
